```python
# Import libraries
import pandas as pd
import numpy as np
```

##  Load Data from Excel File

```python
# Load data from excel file
filename = 'data/dataset.xlsx'
shipments = pd.read_excel(filename, sheet_name= 'shipments')
packages = pd.read_excel(filename, sheet_name= 'packages')
products = pd.read_excel(filename, sheet_name= 'products')
```

### Cleaning DataSet

```python
#Clean strings
columnClean = shipments['courier']

correos = columnClean.str.contains('CORREOS')
publicCorreos = columnClean.str.contains('publicCorreos')
envialia = columnClean.str.contains('ENVIALIA')
mrw = columnClean.str.contains('MRW') 
seur = columnClean.str.contains('SEUR') 
zeleris = columnClean.str.contains('zelerisCustom')
glsNew = columnClean.str.contains('glsNew')
glsShipIt = columnClean.str.contains('glsShipIt')
dhlParcel = columnClean.str.contains('dhlParcel')

shipments['courier'] = np.where(correos, 'correos',
                            np.where(publicCorreos, 'correos',
                                 np.where(envialia, 'envialia',
                                     np.where(mrw, 'mrw',
                                        np.where(seur, 'seur',
                                            np.where(zeleris, 'zeleris',
                                                np.where(glsNew, 'gls',
                                                    np.where(glsShipIt, 'gls',
                                                        np.where(dhlParcel, 'dhl',
                                                            columnClean.str.replace(' ', ''))))))))))
                                                            
#Convert Fields to DateTime
shipments[["createdAt", "deliverDate", "estimatedDeliverDate", "pickupDate", "processDate"]] = shipments[["createdAt", "deliverDate", "estimatedDeliverDate", "pickupDate", "processDate"]].apply(pd.to_datetime)
```

## AVERAGE DELIVERY TIME PER COURIER

```python
#Average delivery time per courier
shipments['deliveryTime'] = shipments['deliverDate'] - shipments['createdAt']

#isolating important columns
avg_delivery_time_couriers = shipments[['courier', 'deliveryTime']]

#formatting to days
avg_delivery_time_couriers['deliveryTime'] = avg_delivery_time_couriers['deliveryTime'].dt.days

#average calculation
avg_delivery_time_couriers = avg_delivery_time_couriers.groupby(['courier']).mean().round(0)

#print DELIVERY AVERAGE BY COURIER
avg_delivery_time_couriers
```

## AVERAGE DELIVERY TIME PER SHIPPING METHOD

```python
#Average delivery time per shipping method
#Set option to show all the rows
pd.set_option('display.max_rows', None)

#isolating important columns
avg_delivery_time_shipping_methods = shipments[['method', 'deliveryTime']]

#formatting to days
avg_delivery_time_shipping_methods['deliveryTime'] = avg_delivery_time_shipping_methods['deliveryTime'].dt.days

#average calculation
avg_delivery_time_shipping_methods = avg_delivery_time_shipping_methods.groupby(['method']).mean().round(0)

#print DELIVERY AVERAGE BY COURIER
avg_delivery_time_shipping_methods
```

## AVERAGE PRODUCTS PER ORDER

```python
#Product count
productsList = pd.DataFrame({'original': packages['products']})
productsList.products = productsList.original.str.split(',')
productsList.count = productsList.products.apply(len)

#Packages count
packagesList = pd.DataFrame({'original': shipments['packages']})
packagesList.packages = packagesList.original.str.split(',')
packagesList.count = packagesList.packages.apply(len)

#Creating Avg
avgProductsPerPackage = productsList.count.mean()
avgPackagesPerShipment = packagesList.count.mean()

#Creating Avg Per Shipmennt
avgProductsPerShipment = avgProductsPerPackage * avgPackagesPerShipment

#AVG PRODUCTS PER SHIPMENT
avgProductsPerShipment.round(0)

```
